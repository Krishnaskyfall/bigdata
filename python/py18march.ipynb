```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as plt
x=np.linspace(0,10*np.pi,100)
ysin=np.sin(x)
ycos=np.cos(x)
print
fig=plt.figure()
plt.plot(x,ysin)
plt.plot(x,ycos)

```

```python
fig=plt.figure(figsize=(20,5))
x=np.linspace(0,10*np.pi,100)
ysin=np.sin(x)
ycos=np.cos(x)
plt.plot(x,ysin,color="yellow",linestyle="dashed",label="Sin")
plt.plot(x,ycos,color="g",linestyle="-",label="Cosine")
plt.xlim(0,2*np.pi)
plt.ylim(-1.25,1.25)
plt.xlabel("Time")
plt.ylabel("Response")
plt.title("Trignometric Functions")
plt.legend()
plt.show()
```
